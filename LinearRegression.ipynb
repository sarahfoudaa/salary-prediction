#load processed data

```python
import pandas as pd
X_train = pd.read_csv('/content/drive/MyDrive/machine_learning_projects/salary_prediction/data/selected/X_train.csv')
y_train = pd.read_csv('/content/drive/MyDrive/machine_learning_projects/salary_prediction/data/selected/y_train.csv')
```

```python
from custom_funcs import  splitModeling
X_train, X_test, y_train, y_test = splitModeling(X_train,y_train,0.1)
```

#linear regression

```python
#%%writefile custom model
from sklearn.linear_model import LinearRegression
import argparse
import os
import sys

class CustomLinearRegrresion:
  def __init__(self):
    self.model = LinearRegression()

  def train(self, X, y):
    self.model.fit(X,y)

  def predict(self, X):
    return self.model.predict(X)


```

```python
model = CustomLinearRegrresion()
model.train(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, predictions)
```
